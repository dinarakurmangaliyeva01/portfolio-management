```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas_datareader
import datetime as dt
```

```python
import pandas_datareader.data as web
```

```python
#get closing value of MSFT
start = dt.datetime(2015, 1, 1, 0, 0)
end = dt.datetime.now()
MSFT_df = web.DataReader("MSFT", 'yahoo', start, end)
MSFT_df.reset_index(inplace=True)
MSFT_df.set_index("Date", inplace=True)
```

```python
#get closing value of HSBS
HSBC_df = web.DataReader("HSBC", 'yahoo', start, end)
HSBC_df.reset_index(inplace=True)
HSBC_df.set_index("Date", inplace=True)
```

```python
#get closing value of Coca Cola
KO_df = web.DataReader("KO", 'yahoo', start, end)
KO_df.reset_index(inplace=True)
KO_df.set_index("Date", inplace=True)
```

```python
from scipy import stats

import matplotlib.pyplot as plt
%matplotlib inline


KO_df['Close'].plot(label='KO')
MSFT_df['Close'].plot(label='MSFT')
plt.legend()
```

```python
import warnings
warnings.filterwarnings('ignore')

KO_df['Cumulative'] = KO_df['Close']/KO_df['Close'].iloc[0]
MSFT_df['Cumulative'] = MSFT_df['Close']/MSFT_df['Close'].iloc[0]
```

```python
KO_df['Cumulative'].plot(label='KO',figsize=(10,8))
MSFT_df['Cumulative'].plot(label='MSFT')
plt.legend()
plt.title('Cumulative Return')
```
